```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
```

```python
text_file=open("value.txt", "r")
values = []
for line in text_file:
    values.append(line.strip("\n").split(","))
text_file.close() 
#print(values)
#print(len(values))
values = np.array(values)
values = values.astype(np.float)
```

```python
# number of detector
a= (values[:,0]+values[:,1])/2 
b= (values[:,5]+values[:,6])/2 
c= (values[:,10]+values[:,11])/2 
d= (values[:,15]+values[:,16])/2 
numdetect = np.column_stack((a, b,c,d))
```

```python
# number of inliers after check
a= values[:,3]
b= values[:,8]
c= values[:,13]
d= values[:,18]
numinlier = np.column_stack((a, b,c,d))
print(numinlier)
```

```python
# number of time cost
a= values[:,4]
b= values[:,9]
c= values[:,14]
d= values[:,19]
numtime = np.column_stack((a, b,c,d))
print(numtime)
```

```python
x = np.array([1,2,3,4])
my_xticks = ['s1','s2','s3','s4']
fig = plt.figure()
ax =plt.subplot(1,1,1)
plt.xticks(x,my_xticks)
plt.xlabel("# of scenarios")
plt.ylabel("# of detected features")
ax.plot(x,numdetect[0],'o-',label='FAST-patch') 
ax.plot(x,numdetect[1],'+-',label='FAST-SIFT')
ax.plot(x,numdetect[2],'D-',label='Harris-SURF') 
ax.plot(x,numdetect[3],'h-',label='FAST-ORB')
ax.plot(x,numdetect[4],'H-',label='ORB-ORB') 
ax.plot(x,numdetect[5],'p-',label='SURF-LIOP')
ax.plot(x,numdetect[6],'8-',label='SIFT-SIFT') 
ax.plot(x,numdetect[7],'s-',label='SURF-BRISK')
ax.plot(x,numdetect[8],'*-',label='SURF-BRIEF') 
ax.plot(x,numdetect[9],'v-',label='SURF-SURF')
ax.plot(x,numdetect[10],'x-',label='SURF-SIFT') 
ax.plot(x,numdetect[11],'>-',label='MSER-SIFT')
ax.plot(x,numdetect[12],'d-',label='BRISK-BRISK')
ax.plot(x,numdetect[13],'^-',label='BRISK-FREAK') 
ax.plot(x,numdetect[14],'_-',label='FAST-BRIEF')

ax.legend(bbox_to_anchor=(1.23,1),prop={'size':6})
plt.show()
```
